```python
# STEP 1: Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Display settings and inline plotting
sns.set(style="whitegrid")
%matplotlib inline

# STEP 2: Confirm Current Folder Contents
print("Files in current directory:")
print(os.listdir())  # Check for correct file names

# STEP 3: Load the Data
customer_df = pd.read_csv('client_data (1).csv')
pricing_df = pd.read_csv('price_data (1).csv')

# STEP 4: Preview the Data
print("\nCustomer Data (first 5 rows):")
print(customer_df.head())

print("\nPricing Data (first 5 rows):")
print(pricing_df.head())

# STEP 5: Data Types and Missing Values
print("\nCustomer Data Types:\n", customer_df.dtypes)
print("\nPricing Data Types:\n", pricing_df.dtypes)

print("\nMissing Values in Customer Data:\n", customer_df.isnull().sum())
print("\nMissing Values in Pricing Data:\n", pricing_df.isnull().sum())

# STEP 6: Descriptive Statistics
print("\nCustomer Data Summary:\n", customer_df.describe())
print("\nPricing Data Summary:\n", pricing_df.describe())

# Unique value counts for object (categorical) columns
for col in customer_df.select_dtypes(include='object').columns:
    print(f"{col}: {customer_df[col].nunique()} unique values")

# STEP 7: Merge Datasets (Optional  only if you know the matching column)
#  Adjust 'Customer_ID' to the correct common column name
# merged_df = pd.merge(customer_df, pricing_df, on='Customer_ID')

# STEP 8: Visualization Examples

# Histogram: Example numeric column (change if needed)
if 'Forecasted_Usage' in customer_df.columns:
    sns.histplot(customer_df['Forecasted_Usage'], kde=True)
    plt.title('Distribution of Forecasted Usage')
    plt.xlabel('Forecasted Usage')
    plt.ylabel('Frequency')
    plt.show()

# Heatmap of numeric columns
plt.figure(figsize=(10, 6))
sns.heatmap(customer_df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix - Customer Data')
plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(pricing_df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix - Pricing Data')
plt.show()
```
