```python
import pandas as pd

# Load the cleaned dataset
df = pd.read_csv('clean_data_after_eda.csv')

# View the first few rows
print("Initial data preview:")
print(df.head())

# Feature 1: Total variable price
df['total_variable_price'] = (
    df['var_6m_price_off_peak'] +
    df['var_6m_price_peak'] +
    df['var_6m_price_mid_peak']
)

# Feature 2: Total fixed price (optional but useful)
df['total_fixed_price'] = (
    df['var_6m_price_off_peak_fix'] +
    df['var_6m_price_peak_fix'] +
    df['var_6m_price_mid_peak_fix']
)

# Feature 3: Price ratio (variable to fixed)
df['price_ratio_var_to_fix'] = df['total_variable_price'] / (df['total_fixed_price'] + 1e-5)

# View new features
print("\nNewly engineered features:")
print(df[['total_variable_price', 'total_fixed_price', 'price_ratio_var_to_fix']].head())

# Save the updated dataset
df.to_csv('engineered_data.csv', index=False)
print("\nSaved 'engineered_data.csv' with new features.")
```
